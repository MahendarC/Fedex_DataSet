```python
#Z_Score for outliers detection   ---p values   ---scalling

from scipy import stats

z_scores = stats.zscore(df[['Actual_Shipment_Time', 'Planned_TimeofTravel', 'Shipment_Delay', 'Distance']])
outliers = (z_scores > 3.5).sum(axis=0)
print(outliers)
```

```python
num_cols = ['Actual_Shipment_Time', 'Planned_TimeofTravel', 'Shipment_Delay', 'Distance']


#compute z-score
z_scores = stats.zscore(df[num_cols])

#Filter data to remove outliers (keep only rows where all z-score are < 3 or 2.5 or 3.5)
df_cleaned = df[(abs(z_scores) < 3.5).all(axis=1)]


#to know 
print(f'Original Data Shape: {df.shape}')
print(f'Cleaned Data Shape: {df_cleaned.shape}')

#how many outliers are 
print('Outliers removed:', {len(df)-len(df_cleaned)})
```

```python
''''''
#----- if the data is skewed, consider IQR

num_cols = ['Actual_Shipment_Time', 'Planned_TimeofTravel', 'Shipment_Delay', 'Distance']

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR


    filtered_data = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_data[column]

for col in num_cols:
    plt.figure(figsize=(12, 5))

    filtered_col = remove_outliers(df, col)

    min_val, max_val = filtered_col.min(), filtered_col.max()


    plt.subplot(1, 2, 1)
    sns.histplot(filtered_col, bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xlim(min_val, max_val)

    plt.subplot(1, 2, 2)
    sns.boxplot(x=filtered_col)
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.xlim(min_val, max_val)

    plt.show()
 
```
