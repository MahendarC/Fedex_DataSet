```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, PowerTransformer
from scipy.stats.mstats import winsorize

# handle outliers using IQR
Q1 = df['Shipment_Delay'].quantile(0.25)
Q3 = df['Shipment_Delay'].quantile(0.75)
IQR = Q3 - Q1
df_filtered = df[(df['Shipment_Delay'] >= (Q1 - 1.5 * IQR)) & (df['Shipment_Delay'] <= (Q3 + 1.5 * IQR))]

df_filtered['Travel_Speed'] = df_filtered['Distance'] / df_filtered['Planned_TimeofTravel']
df_filtered['Delay_Percentage'] = df_filtered['Shipment_Delay'] / df_filtered['Planned_TimeofTravel']

#scaler = RobustScaler()
#scaler = StandardScaler()   
scaler = MinMaxScaler()

df_scaled = scaler.fit_transform(df_filtered[['Actual_Shipment_Time', 'Planned_TimeofTravel', 'Shipment_Delay', 'Distance', 'Travel_Speed', 'Delay_Percentage']])
df_scaled = pd.DataFrame(df_scaled, columns=['Actual_Shipment_Time', 'Planned_TimeofTravel', 'Shipment_Delay', 'Distance', 'Travel_Speed','Delay_Percentage'])

#power Transformation for skewed Data
pt = PowerTransformer(method='yeo-johnson')
df_scaled[['Shipment_Delay']] = pt.fit_transform(df_scaled[['Shipment_Delay']])

#plot histogram after scaling
plt.figure(figsize=(12, 6))
for col in df_scaled.columns:
    sns.histplot(df_scaled[col], bins=30, kde=True, label=col, alpha=0.6)

plt.title("Distribution of Processed Numerical Features")
plt.legend()
plt.show()

print(df_scaled.head())



'''    #RobustScaler
Actual_Shipment_Time  Planned_TimeofTravel  Shipment_Delay  Distance  \
0              0.938442                0.5000        0.988763  0.362916   
1             -0.630653                0.4375        1.702683  0.362916   
2             -0.788945               -0.2500        0.988763 -0.104596   
3             -0.414573               -0.2500       -0.470276 -0.104596   
4              0.859296                0.0625        1.999960  0.169572   

   Travel_Speed  Delay_Percentage  
0      0.117501          0.954302  
1      0.214446          2.093359  
2      0.285260          1.475527  
3      0.285260         -0.479067  
4      0.420826          3.359303 '''



'''
# Standard scaler
Actual_Shipment_Time  Planned_TimeofTravel  Shipment_Delay  Distance  \
0              1.495573              0.300268        1.012511  0.142430   
1             -1.154175              0.228692        1.697676  0.142430   
2             -1.421483             -0.558646        1.012511 -0.380379   
3             -0.789277             -0.558646       -0.484421 -0.380379   
4              1.361919             -0.200765        1.974467 -0.073783   

   Travel_Speed  Delay_Percentage  
0      0.110369          0.468111  
1      0.174971          1.239008  
2      0.222160          0.820868  
3      0.222160         -0.501973  
4      0.312497          2.095781'''
```
